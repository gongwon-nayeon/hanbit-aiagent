### 6.2.2 실습 랭체인에서 도구 호출 사용하기

```python
from langchain.tools import tool

@tool
def add(a: int, b: int) -> int:
    """Adds a and b.

    Args:
        a: first int
        b: second int
    """
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiplies a and b.

    Args:
        a: first int
        b: second int
    """
    return a * b

tools = [add, multiply]
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o")
llm_with_tools = llm.bind_tools(tools)
```

```python
query = "3 곱하기 5 는 뭔가요? 그리고 2 더하기 4 는 뭔가요?"

response = llm_with_tools.invoke(query)
response
```

```python
response.tool_calls
```

```python
query = "안녕하세요."

response = llm_with_tools.invoke(query)
response
```

```python
response.tool_calls
```

### Tavily Search 적용하기

```python
from langchain_openai import ChatOpenAI

tool = TavilySearch(max_results=2)
tools = [tool]

llm = ChatOpenAI(model="gpt-4o")
llm_with_tools = llm.bind_tools(tools)
```
